```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Flatten, Reshape
from tensorflow.keras.models import Sequential
from matplotlib import pyplot as plt
```

```python
# Load the MNIST dataset
(train_x, train_y), (test_x, test_y) = keras.datasets.mnist.load_data(path='mnist.npz')
```

```python
# Print the number of training samples
print(len(train_x))
```

```python
print(train_x[0].shape)
```

```python
# Plot the first training sample
plt.imshow(train_x[0], cmap='gray')
plt.show()
```

```python
# Reshape the data to be (num_samples, 28*28)
train_x = train_x.reshape(-1, 28*28).astype('float32') / 255
test_x = test_x.reshape(-1, 28*28).astype('float32') / 255
```

```python
# Define the encoder
latent_dim = 64
encoder = Sequential([
    Flatten(),
    Dense(latent_dim, activation='linear')
])
```

```python
# Define the decoder
decoder = Sequential([
    Dense(784, activation='linear'),
    Reshape((28*28,)),
])
```

```python
# Define the autoencoder
autoencoder = Sequential([encoder, decoder])
```

```python
# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
autoencoder.fit(train_x, train_x, epochs=10, batch_size=256, validation_data=(test_x, test_x))
```

```python
# Evaluate the model (this will not give accuracy, but reconstruction loss)
loss = autoencoder.evaluate(test_x, test_x)
print(f"Reconstruction loss: {loss}")
```

```python
# Plot the original and reconstructed test sample
plt.imshow(test_x[0].reshape(28, 28), cmap='gray')
plt.title("Original")
plt.show()

reconstructed_img = autoencoder.predict(test_x[0].reshape(1, 28*28))
plt.imshow(reconstructed_img.reshape(28, 28), cmap='gray')
plt.title("Reconstructed")
plt.show()
```

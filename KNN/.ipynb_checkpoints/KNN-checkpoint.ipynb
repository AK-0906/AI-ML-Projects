```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import joblib
from matplotlib import pyplot as plt
```

```python
Raw_data = pd.read_csv('Customer_Behaviour.csv')
print(Raw_data.columns)
```

```python
label_encoder = LabelEncoder()
Raw_data['Gender'] = label_encoder.fit_transform(Raw_data['Gender'])
data = Raw_data.drop('User ID', axis = 1)
```

```python
Test_data, Train_data = train_test_split(data, train_size = 0.25, test_size = 0.75, random_state = 1024)
print(len(Train_data))
```

```python
X_train = Train_data.drop('Purchased', axis=1)
y_train = Train_data['Purchased']
X_test = Test_data.drop('Purchased', axis=1)
y_test = Test_data['Purchased']
```

```python
X_axis = []
Y_axis = []
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    X_axis.append(i)
    Y_axis.append(accuracy)
print(Y_axis)
```

```python
plt.plot(X_axis, Y_axis)
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Classifier Accuracy with Varying k')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
joblib.dump(knn, 'Output_weight.joblib')
```

```python
model = joblib.load('Output_weight.joblib')
```

```python
print(X_train)
print(y_train)
```

```python
Total = 0
Correct = 0
for i in range(300):
    x = [X_test.iloc[i].values]
    y = y_test.iloc[i]
    y_pred = model.predict(x)
    if y == y_pred:
        Correct += 1
    Total += 1
```

```python
print(f'Correctly classified: {Correct}/{Total}')
```

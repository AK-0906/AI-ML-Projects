```julia
using Transformers
using Images
```

```julia
# Load the image
image_path = "Backend/Images/download.jpg"
image = load(image_path)
```

```julia
# Load the model
saved_model_directory = "./saved_model_pipe"
pipe = Transformers.pipeline("image-to-text", model=saved_model_directory)
```

```julia
# Make a prediction
prediction = pipe(image)
```

```julia
# Print the prediction
println(prediction)
```
